# Exercício 1

Num jogo de snooker, um jogador vai jogar a bola azul contra a amarela de modo a que a bola amarela caia na bolsa do canto direito. Para tal a bola amarela tem de sair da colisão com a bola azul a fazer um ângulo de $\phi = 35^\circ$.

Não considere o atrito e a resistência do ar, de modo que a colisão é elástica.

Note: As bolas de snooker têm a mesma massa.

A bola azul bate na amarela com a velocidade de 1 m/s. Nestas condições calcule:

## Parte A

O ângulo $\theta$ que a bola azul faz com a direção inicial.

Aplicando a lei da conservação do momento temos que:

$$
\vec{p_{1,i}} + \vec{p_{2,i}} = \vec{p_{1,f}} + \vec{p_{2,f}}
$$

A bola amarela encontra-se em repouso no início, logo $\vec{p_{2,i}} = \vec{0}$, o que simplifica a expressão da conservação do momento:

$$
\vec{p_{1,i}} = \vec{p_{1,f}} + \vec{p_{2,f}}
$$

Se aplicarmos agora o quadrado aos dois lados da equação, obtemos a seguinte equação:

$$
p_{1,i}^2 = p_{1,f}^2 + 2 \vec{p_{1,f}} \cdot \vec{p_{2,f}} + p_{2,f}^2
$$

Se utilizarmos a definição de momento como $\vec{p} = m \vec{v}$, obtemos a seguinte equação:

$$
\begin{align*}
&m^2 v_{1,i}^2 = m^2 v_{1,f}^2 + 2 \left( m \vec{v_{1,f}} \right) \cdot \left( m \vec{v_{2,f}} \right) + m^ 2 v_{2,f}^2 \\
\Leftrightarrow
&m^2 v_{1,i}^2 = m^2 v_{1,f}^2 + 2 m^2 \vec{v_{1,f}} \cdot \vec{v_{2,f}} + m^ 2 v_{2,f}^2 \\
\Leftrightarrow
&m^2 v_{1,i}^2 = m^2 \left( v_{1,f}^2 + 2 \vec{v_{1,f}} \cdot \vec{v_{2,f}} + v_{2,f}^2 \right) \\
\Leftrightarrow
&v_{1,i}^2 = v_{1,f}^2 + 2 \vec{v_{1,f}} \cdot \vec{v_{2,f}} + v_{2,f}^2 \\
\Leftrightarrow
&1 = v_{1,f}^2 + 2 \vec{v_{1,f}} \cdot \vec{v_{2,f}} + v_{2,f}^2
\end{align*}
$$

Como a colisão elástica temos também a conversão da energia mecânica, neste caso como não existe energia potencial, a energia mecânica será igual a energia cinêtica:

$$
\begin{align*}
&\frac{1}{2} m v_{1,i}^2 + \frac{1}{2} m v_{2,i}^2 = \frac{1}{2} m v_{1,f}^2 + \frac{1}{2} m v_{2,f}^2 \\
\Leftrightarrow
&\frac{1}{2} m \left( v_{1,i}^2 + v_{2,i}^2 \right) = \frac{1}{2} m \left( v_{1,f}^2 +  v_{2,f}^2 \right) \\
\Leftrightarrow
&v_{1,i}^2 + v_{2,i}^2 = v_{1,f}^2 +  v_{2,f}^2 \\
\Leftrightarrow
&1^2 + 0^2 = v_{1,f}^2 +  v_{2,f}^2 \\
\Leftrightarrow
&1 = v_{1,f}^2 + v_{2,f}^2 \\
\Leftrightarrow
&v_{1,f}^2 = 1 - v_{2,f}^2
\end{align*}
$$

Substituindo na equação da conservação do momento temos que:

$$
\begin{align*}
&1 = v_{1,f}^2 + 2 \vec{v_{1,f}} \cdot \vec{v_{2,f}} + v_{2,f}^2 \\
\Leftrightarrow
&1 = 1 - v_{2,f}^2 + 2 \vec{v_{1,f}} \cdot \vec{v_{2,f}} + v_{2,f}^2 \\
\Leftrightarrow
&0 = 2 \vec{v_{1,f}} \cdot \vec{v_{2,f}} \\
\Leftrightarrow
&0 = \vec{v_{1,f}} \cdot \vec{v_{2,f}}
\end{align*}
$$

O produto escalar entre dois vetores é definido como as magnitudes dos dois vetores multiplicados pelo cosseno do ângulo entre os vetores, logo:

$$
0 = \vec{v_{1,f}} \cdot \vec{v_{2,f}}
\Leftrightarrow
0 =
\left\lvert \left\lvert v_{1,f} \right\rvert \right\rvert
\left\lvert \left\lvert v_{2,f} \right\rvert \right\rvert
\cos \left( \theta + \phi \right)
$$

O produto escalar só será zero quando as magnitudes dos vetores forem zero, o que só pode acontecer quando o ângulo entre os vetores é zero (a bola azul bate diretamente na bola amarela, saindo na mesma direção). Como o ângulo da bola amarela não é nulo esta situação não pode acontecer, o que só deixa a opção do cosseno ser zero, ou seja o ângulo entre os vetores é de 90 graus.

O que nos dá $\theta = 90^\circ - 35^\circ = 55^\circ$.

## Parte B

Como acabamos de ver na alínea anterior, o ângulo entre as duas bolas será sempre 90 graus.

## Parte C

A velocidade da bola azul depois da colisão.

Como já vimos, segundo as leis da conservação do momento, temos que:

$$
\vec{p_{1,i}} = \vec{p_{1,f}} + \vec{p_{2,f}}
$$

E aplicando a definição do momento:

$$
m\vec{v_{1,i}} = m\vec{v_{1,f}} + m\vec{v_{2,f}}
\Leftrightarrow \vec{v_{1,i}} = \vec{v_{1,f}} + \vec{v_{2,f}}
$$

Substituindo os valores e as expressões dos vetores:

$$
(1, 0) = 
\left( v_{1,f} \cos(55^\circ), v_{1,f} \sin(55^\circ) \right)
+ \left( v_{2,f} \cos(35^\circ), -v_{2,f} \sin(35^\circ) \right)
$$

O que pode ser resolvido como um sistema de duas equações para cada componente do vetor:

$$
\begin{align*}
    &\begin{cases}
      1 = v_{1,f} \cos(55^\circ) + v_{2,f} \cos(35^\circ) \\
      0 = v_{1,f} \sin(55^\circ) - v_{2,f} \sin(35^\circ)
    \end{cases} \\
    \Leftrightarrow
    &\begin{cases}
      1 = v_{1,f} \cos(55^\circ) + v_{2,f} \cos(35^\circ) \\
      v_{1,f} = v_{2,f} \frac{\sin(35^\circ)}{\sin(55^\circ)}
    \end{cases} \\
    \Leftrightarrow
    &\begin{cases}
      1 = v_{2,f} \frac{\sin(35^\circ)\cos(55^\circ)}{\sin(55^\circ)} + v_{2,f} \cos(35^\circ) \\
      v_{1,f} = v_{2,f} \frac{\sin(35^\circ)}{\sin(55^\circ)}
    \end{cases} \\
    \Leftrightarrow
    &\begin{cases}
      1 = v_{2,f} \left( \frac{\sin(35^\circ)\cos(55^\circ)}{\sin(55^\circ)} + \cos(35^\circ) \right) \\
      v_{1,f} = v_{2,f} \frac{\sin(35^\circ)}{\sin(55^\circ)}
    \end{cases} \\
    \Leftrightarrow
    &\begin{cases}
      v_{2,f} = \frac{1}{\frac{\sin(35^\circ)\cos(55^\circ)}{\sin(55^\circ)} + \cos(35^\circ)} \approx 0.819 \\
      v_{1,f} = v_{2,f} \frac{\sin(35^\circ)}{\sin(55^\circ)}
    \end{cases} \\
    \Leftrightarrow
    &\begin{cases}
      v_{2,f} \approx 0.819 \\
      v_{1,f} \approx 0.819 \frac{\sin(35^\circ)}{\sin(55^\circ)} \approx 0.573
    \end{cases}
\end{align*}
$$

Ou seja a velocidade da bola azul após a colisão é 0.573 m/s.

# Exercício 3

Uma mola exerce uma força $F_x = -k x(t)$, em que $k$ é a constante elástica da mola, num corpo de massa m. Considere $k = 1 \, \text{N/m}$ e m = $1 \, \text{kg}$.

## Parte A

Mostre que a lei do movimento $x(t) = A \cos(\omega t + \phi)$, com $\omega = \sqrt{k/m}$, é solução da equação dinâmica de Newton do sistema mola-corpo. Qual a lei de velocidade do corpo ligado à mola, em que A e $\phi$ são constantes?

Para verificar basta calcular a derivada de segunda ordem da lei do movimento que nos dará a aceleração, a lei da velocidade será a derivada de primeira ordem da lei do movimento.

$$
v = \frac{dx}{dt} = \left( A \cos(\omega t + \phi) \right)' = - A \omega \sin(\omega t + \phi)
$$

$$
\begin{align*}
a &= \frac{dv}{dt} = \left( - A \omega \sin(\omega t + \phi) \right)' \\
&= - A \omega^2 \cos(\omega t + \phi) \\
&= - A \frac{k}{m} \cos(\omega t + \phi)
\end{align*}
$$

Aplicando $F = ma$, verificamos que é solução:

$$
\begin{align*}
F &= - m\frac{k}{m} A \cos(\omega t + \phi) \\
&= - k A \cos(\omega t + \phi) = - k x(t)
\end{align*}
$$

## Parte B e C

b) Calcule numericamente a lei da velocidade e compare com o resultado analítico. Qual o método numérico que escolhe? Considere nula a velocidade inicial e a posição inicial 4 m.

c) Calcule numericamente a lei do movimento nas condições da alínea anterior e compare com o resultado analítico.

Vamos escolher o método de Euler-Cromer, pois este produz espaços de fase fechados, quando comparado com o método de Euler para oscilações.

```python
# Dados
k = 1
m = 1
# Parâmetros
dt = 0.001 # δt - tamanho do passo
t0 = 0 # Tempo inicial
tf = 30.0 # Tempo final
x0 = 4 # Posição inicial
v0 = 0 # Velocidade inicial

# Número de passos/iterações
#
# + 0.1 para garantir que não há arrendodamentos
# para baixo
n = int((tf-t0) / dt + 0.1)

t = np.zeros(n + 1) # Tempo
x = np.zeros(n + 1) # Posição
v = np.zeros(n + 1) # Velocidade
a = np.zeros(n + 1) # Aceleração

# Valores inicias
v[0] = v0
t[0] = t0
x[0] = x0

for i in range(n):
  a[i] = -(k/m) * x[i] # Força a dividir por massa
  v[i + 1] = v[i] + a[i] * dt
  x[i + 1] = x[i] + v[i + 1] * dt 
  t[i + 1] = t[i] + dt
```

Precisamos agora de calcular $A$ e $\phi$, já fizemos isto no exercício 2.a). Os valores são $A = 4$ e $\phi = 0$.

```python
x_analitico = 4 * np.cos(t)

plt.plot(t, x, label="Númerico")
plt.plot(t, x_analitico, label="Analítico")
plt.ylabel("x (m)")
plt.xlabel("t (s)")
plt.title("Sistema Mola-Corpo")
plt.legend()
plt.grid()
plt.show()
```

Os resultados analíticos e númericos para a lei do movimento são aproximadamente iguais.

```python
v_analitico = - 4 * np.sin(t)

plt.plot(t, v, label="Númerico")
plt.plot(t, v_analitico, label="Analítico")
plt.ylabel("v (m/s)")
plt.xlabel("t (s)")
plt.title("Sistema Mola-Corpo")
plt.legend()
plt.grid()
plt.show()
```
