# Exercício 1

Um vetor a 2 dimensões tem as coordenadas (3, 4).

a) Qual a sua intensidade ou comprimento?

$$
\left\lvert (3,  4) \right\rvert = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
$$

b) Qual o vetor unitário correspondente?

Para calcular o vetor unitário, precisamos de multiplicar o vetor pelo o inverso da intensidade.

$$
\frac{1}{5} (3, 4) = \left(\frac{3}{5}, \frac{4}{5}\right)
$$

c) Qual o vetor $2\times(3, 4)$? Qual o seu comprimento ou módulo?

$$
2\times(3, 4) = (6, 8)
$$

$$
\left\lvert (6,  8) \right\rvert = \sqrt{6^2 + 8^2} = \sqrt{36 + 64} = \sqrt{100} = 10
$$

ou

$$
\left\lvert 2\times(3, 4) \right\rvert = 2 \left\lvert (3, 4) \right\rvert = 2 \times 5 = 10
$$

# Exercício 2

Dois vetores são (1, 2) e (-2, 3). Qual o seu produto escalar e qual o ângulo entre os dois
vetores?

$$
(1, 2) \cdot (-2, 3) = 1 \times -2 + 2 \times 3 = -2 + 6 = 4
$$

Para calcular o ângulo

$$
\begin{align*}
\cos \theta 
= \frac{(1, 2) \cdot (-2, 3)}{\left\lvert (1, 2) \right\rvert \left\lvert (-2, 3) \right\rvert}
&\Leftrightarrow \theta 
= \cos^{-1} \frac{(1, 2) \cdot (-2, 3)}{\left\lvert (1, 2) \right\rvert \left\lvert (-2, 3) \right\rvert} \\
&\Leftrightarrow \theta = \cos^{-1} \frac{4}{\sqrt{5} \sqrt{13}} \\
&\Leftrightarrow \theta = \cos^{-1} \frac{4}{\sqrt{65}} \approx 1.0517 \; (\text{rad.})
\end{align*}
$$

# Exercício 3

Encontre um vetor perpendicular ao vetor (3, 4), no espaço a 2D. Note que o produto
escalar de dois vetores perpendiculares é nulo.

Sabemos que um vetor perpendicular a outra têm entre eles um ângulo de $\frac{\pi}{2}$ radianos, logo se substituírmos no produto escalar obtemos os vetores perpendiculares.

$$
\begin{align*}
\cos \frac{\pi}{2}
= \frac{(3, 4) \cdot (a, b)}{\left\lvert (3, 4) \right\rvert \left\lvert (a, b) \right\rvert}
&\Leftrightarrow 0 = \frac{3a + 4b}{5 \sqrt{a^2 + b^2}} \\
&\Leftrightarrow 0 = 3a + 4b \\
&\Leftrightarrow -3a = 4b \\
&\Leftrightarrow a = -\frac{4}{3}b
\end{align*}
$$

Logo desde que a igualdade se mantenha o vetor é perpendicular a (3, 4). Usando por exemplo $b = 3$, ficamos com (-4, 3) que é perpendicular a (3, 4).

```python
plt.arrow(0, 0, 3, 4, color="r", width=0.1)
plt.arrow(0, 0, -4, 3, color="b", width=0.1)
plt.axis('equal')
plt.show()
```

# Exercício 4

Uma força tem como coordenada segundo OX $F_x = 2.0N$ e faz um ângulo de 60º com o eixo
dos OX. Qual a intensidade (ou módulo) da força?

Como $F_x$ têm coordenada de $x = 2.0N$ e sabemos o ângulo que a força faz com o eixo, podemos
através do $\cos$ calcular a intensidade da força, pois esta será a hipotenusa de um triângulo retângulo formado entre as coordenadas de $x$ e as de $y$.

$$
\begin{align*}
\cos 60^\circ = \frac{2}{\lvert F \rvert}
&\Leftrightarrow \lvert F \rvert = \frac{2}{\cos 60^\circ} \\
&\Leftrightarrow \lvert F \rvert = \frac{2}{\frac{1}{2}} \\
&\Leftrightarrow \lvert F \rvert = 4
\end{align*}
$$

```python
from matplotlib.patches import Arc

a = Arc((0, 0), 2, 2, theta2=60, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, 2, 0, color="b", width=0.05)
plt.arrow(0, 0, 2, np.sqrt(12), color="r", width=0.1)
plt.axis('equal')
plt.show()
```

# Exercício 5

Uma força de intensidade $\lvert \vec{F} \rvert = 5.00 N$ faz um ângulo $\theta$ com o eixo positivo OX. Quais as coordenadas ($F_x$, $F_y$) da força, quando o ângulo for:

As coordenadas de uma força podem ser calculadas utilizando as seguintes expressões:

$$
F_x = \lvert \vec{F} \rvert \cos \theta \\
F_y = \lvert \vec{F} \rvert \sin \theta
$$

a) $\theta = \frac{\pi}{2}$

$$
F_x = \lvert \vec{F} \rvert \cos \frac{\pi}{2} = 5 \times 0 = 0 \\
F_y = \lvert \vec{F} \rvert \sin \frac{\pi}{2} = 5 \times 1 = 5
$$

```python
from matplotlib.patches import Arc

a = Arc((0, 0), 2, 2, theta2=90, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, 0, 5, color="b", width=0.05)
plt.axis('equal')
plt.show()
```

b) $\theta = 60^\circ$

$$
F_x = \lvert \vec{F} \rvert \cos 60^\circ = 5 \times \frac{1}{2} = 2.5 \\
F_y = \lvert \vec{F} \rvert \sin 60^\circ = 5 \times \frac{\sqrt{3}}{2} = \frac{5\sqrt{3}}{2}
$$

```python
from matplotlib.patches import Arc

a = Arc((0, 0), 2, 2, theta2=60, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, 2.5, 5 * np.sqrt(3) / 2, color="b", width=0.05)
plt.axis('equal')
plt.show()
```

c) $\theta = \frac{-7\pi}{6}$

$$
F_x = \lvert \vec{F} \rvert \cos \frac{-7\pi}{6} = 5 \times \frac{- \sqrt{3}}{2} = -\frac{5\sqrt{3}}{2} \\
F_y = \lvert \vec{F} \rvert \sin \frac{-7\pi}{6} = 5 \times \frac{1}{2} = 2.5
$$

```python
from matplotlib.patches import Arc

angle = (-7 * np.pi) / 6

a = Arc((0, 0), 2, 2, theta2=angle * 180 / np.pi, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, -5 * np.sqrt(3) / 2, 2.5, color="b", width=0.05)
plt.axis('equal')
plt.show()
```

d) $\theta = 310^\circ$

$$
F_x = \lvert \vec{F} \rvert \cos 310^\circ = 5 \times \cos \left( \frac{5}{18} \pi \right) \approx 3.21393805 \\
F_y = \lvert \vec{F} \rvert \sin 310^\circ = 5 \times - \sin \left( \frac{5}{18} \pi \right) \approx -3.83022222
$$

```python
from matplotlib.patches import Arc

a = Arc((0, 0), 2, 2, theta2=310, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, 5 * np.cos(5/18 * np.pi), 5 * -np.sin(5/18 * np.pi), color="b", width=0.05)
plt.axis('equal')
plt.show()
```

# Exercício 6

Uma força tem intensidade $\lvert \vec{F} \rvert = 10.0 N$. Quais as coordenadas ($F_x$, $F_y$) da força, quando o ângulo for:

a) $30^\circ$ com a parte positiva do eixo OX

$$
F_x = \lvert \vec{F} \rvert \cos 30^\circ = 10 \times \frac{\sqrt{3}}{2} = 5\sqrt{3} \\
F_y = \lvert \vec{F} \rvert \sin 30^\circ = 10 \times \frac{1}{2} = 5
$$

```python
from matplotlib.patches import Arc

a = Arc((0, 0), 2, 2, theta2=30, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, 5 * np.sqrt(3), 5, color="b", width=0.05)
plt.axis('equal')
plt.show()
```

b) $10^\circ$ com a parte positiva do eixo OY

Como estamos a fazer o ângulo relativamente ao eixo OY positivo, temos que adicionar $\frac{\pi}{2}$ (ou $90^\circ$) ao ângulo.

$$
F_x = \lvert \vec{F} \rvert \cos \left( \theta + \frac{\pi}{2} \right) = - \lvert \vec{F} \rvert \sin \theta \\
F_y = \lvert \vec{F} \rvert \sin \left( \theta + \frac{\pi}{2} \right) = \lvert \vec{F} \rvert \cos \theta
$$

Substituindo pelo ângulo.

$$
F_x = - \lvert \vec{F} \rvert \sin 10^\circ = - 10 \sin \frac{\pi}{18} \approx - 0.17364818 \\
F_y = \lvert \vec{F} \rvert \cos 10^\circ = 10 \cos \frac{\pi}{18} \approx 0.98480775
$$

```python
from matplotlib.patches import Arc

a = Arc((0, 0), 2, 2, theta1=90, theta2=100, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, -10 * np.sin(np.pi / 18), 10 * np.cos(np.pi / 18), color="b", width=0.05)
plt.axis('equal')
plt.show()
```

c) $30^\circ$ com a parte negativa do eixo OX

Como estamos a fazer o ângulo relativamente ao eixo OX negativo, temos que adicionar $\pi$ (ou $180^\circ$) ao ângulo.

$$
F_x = \lvert \vec{F} \rvert \cos \left( \theta + \pi \right) = - \lvert \vec{F} \rvert \cos \theta \\
F_y = \lvert \vec{F} \rvert \sin \left( \theta + \pi \right) = - \lvert \vec{F} \rvert \sin \theta
$$

Substituindo pelo ângulo.

$$
F_x = - \lvert \vec{F} \rvert \cos 30^\circ = - 10 \times \frac{\sqrt{3}}{2} = - 5\sqrt{3} \\
F_y = - \lvert \vec{F} \rvert \sin 30^\circ = - 10 \times \frac{1}{2} = - 5
$$

```python
from matplotlib.patches import Arc

a = Arc((0, 0), 2, 2, theta1=180, theta2=210, linewidth=4)
plt.gca().add_patch(a)
plt.arrow(0, 0, -5 * np.sqrt(3), -5, color="b", width=0.05)
plt.axis('equal')
plt.show()
```

# Exercício 7

A lei do movimento de um objeto de massa 1 kg é $\vec{r} = (2t, t^2) \; m$

a) Calcule a lei da velocidade.

A lei da velocidade pode ser calculada através da derivada, a derivada de um vetor é igual as derivadas do seus componentes, logo a lei da velocidade é dada por:

$$
\vec{v} = \vec{r}' = \left( (2t)', \left( t^2 \right)' \right) = (2, 2t)
$$

b) Represente graficamente (no `python`) o vetor $\vec{r}$ para os tempos $t = 1,2,3,4$

```python
colors = {1: "r", 2: "g", 3: "b", 4: "y" }

for t in range(1, 5):
    plt.arrow(
        0, 0, # start
        2*t, t**2, # end
        color=colors[t], width=0.05,
        label=f"t = {t}"
    )

plt.legend(loc="upper left")
plt.axis('equal')
plt.show()
```

c) Represente graficamente o vetor de velocidade nos mesmos momentos. O posicionamento deve ser tal que o ponto inicial do vetor de velocidade coincide com a posição da massa.

```python
colors = {1: "r", 2: "g", 3: "b", 4: "y" }

for t in range(1, 5):
    x = 2 * t
    y = t**2
    
    plt.arrow(
        x, y, # start
        x + 2, y + 2 * t, # end
        color=colors[t], width=0.05,
        label=f"t = {t}"
    )

plt.legend(loc="upper left")
plt.axis('equal')
plt.show()
```

# Exercício 8

Uma bola de futebol é pontapeada de modo que roda sobre si própria, o que resulta adicionar a força de Magnus às outras forças. A força de Magnus resulta de o escoamento do ar ser diferente nos dois lados opostos da bola.

Se a rotação for descrita pelo vetor $\vec{w} = (0, 0, 10)$ rad/s e a velocidade for $\vec{v} = (0, 1, 0)$ m/s, qual a força de Magnus, se for definida por $\vec{F}_\text{Magnus} = \frac{1}{2} A \rho_\text{ar} r \vec{w} \times \vec{v}$?

$A = \pi r^2$ é a área da secção de corte da bola, $r$ o raio da bola e $\rho_\text{ar} = 1.225 \; \text{kg/m}^3$ a massa volúmica do ar. O raio da bola de futebol é 11 cm.

Faça o cálculo analiticamente, e depois verifica utilizando a função `numpy.cross()`

Começamos por calcular a área.

$$
A = (0.11)^2\pi = 0.0121\pi
$$

De seguida calculamos o produto vectorial.


$$
\vec{w} \times \vec{v} = \begin{vmatrix}
\vec{i} & \vec{j} & \vec{k} \\
0       & 0       & 10      \\
0       & 1       & 0
\end{vmatrix} = -10\vec{i}
$$

$$
\begin{align*}
\vec{F}_\text{Magnus}
&= \frac{1}{2} \times 0.0121\pi \times 1.225 \times 0.11 \times -10\vec{i} \\
&= \frac{1}{2} \times 0.0121\pi \times 1.225 \times 0.11 \times -10\vec{i} \\
&= 0.0008152375 \times -10\vec{i} \pi \\
&= - 0.008152375 \pi \vec{i} \approx - 0.02561144 \vec{i}
\end{align*}
$$

```python
r = 11 / 100
w = np.array([0, 0, 10])
v = np.array([0, 1, 0])
P_ar = 1.225

A = np.pi * r**2

w_v_cross = np.cross(w, v)

F_magnus = 1/2 * A * P_ar * r * w_v_cross
```

```python
F_magnus_x = F_magnus[0]
F_magnus_y = F_magnus[1]
F_magnus_z = F_magnus[2]

display(Math(r'\vec{{F}}_\text{{Magnus}} = ({:.8f}, {:.8f}, {:.8f})'.format(F_magnus_x, F_magnus_y, F_magnus_z)))
```
