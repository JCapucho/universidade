```python
import numpy as np
import matplotlib.pyplot as plt

POINTS = 10
random = np.genfromtxt("../random.csv", delimiter=",")
x = np.linspace(5, 10, POINTS)
x_fitted = np.linspace(5, 10, 100)
y = 2 * x**5 + random
```

```python
plt.plot(x, y, "go")
```

```python
import os, sys
# Import linreg from utils package
sys.path.append(os.path.abspath(".."))
from utils.linreg import linreg
```

```python
reg = linreg(x, y)
f, ax = plt.subplots()
plt.plot(x, y, "go")
plt.plot(x_fitted, reg.apply(x_fitted), "g")
plt.text(
    0.01, 0.99, f"$r^2={reg.r2}$", ha="left", va="top", transform=ax.transAxes
)
```

```python
logx = np.log(x)
logy = np.log(y)
plt.plot(logx, logy, "go")
```

```python
log_reg = linreg(logx, logy)
f, ax = plt.subplots()
plt.plot(logx, logy, "go")
logx_fitted = np.log(x_fitted)
plt.plot(logx_fitted, log_reg.apply(logx_fitted), "g")
plt.text(
    0.01, 0.99, f"$r^2={log_reg.r2}$", ha="left", va="top", transform=ax.transAxes
)
```

```python
n = log_reg.m
c = np.exp(log_reg.b)
f, ax = plt.subplots()
plt.plot(x, y, "go")
plt.plot(x_fitted, c * x_fitted**n, "g")
plt.text(
    0.01, 0.99, f"$n={n}$\n$c={c}$", ha="left", va="top", transform=ax.transAxes
)
plt.savefig('sample.png')
```
