```python
import numpy as np
import matplotlib.pyplot as plt

POINTS = 10
random = np.genfromtxt("random.csv", delimiter=",")
x = np.linspace(0, 2, POINTS)
x_fitted = np.linspace(0, 2, 100)
y = 2 * np.exp(2 * x) + random
```

```python
plt.plot(x, y, "go")
```

```python
import os, sys
# Import linreg from utils package
sys.path.append(os.path.abspath(".."))
from utils.linreg import linreg
```

```python
reg = linreg(x, y)
f, ax = plt.subplots()
plt.plot(x, y, "go")
plt.plot(x_fitted, reg.apply(x_fitted), "g")
plt.text(
    0.01, 0.99, f"$r^2={reg.r2}$", ha="left", va="top", transform=ax.transAxes
)
```

```python
logy = np.log(y)
plt.plot(x, logy, "go")
```

```python
semilog_reg = linreg(x, logy)
f, ax = plt.subplots()
plt.plot(x, logy, "go")
plt.plot(x_fitted, semilog_reg.apply(x_fitted), "g")
plt.text(
    0.01, 0.99, f"$r^2={semilog_reg.r2}$", ha="left", va="top", transform=ax.transAxes
)
```

```python
n = semilog_reg.m
c = np.exp(semilog_reg.b)
f, ax = plt.subplots()
plt.plot(x, y, "go")
plt.plot(x_fitted, c * np.exp(x_fitted*n), "g")
plt.text(
    0.01, 0.99, f"$\lambda={n}$\n$y_0={c}$", ha="left", va="top", transform=ax.transAxes
)
```
